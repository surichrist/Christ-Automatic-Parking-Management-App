```python
def init_db():
    conn = sqlite3.connect('vehicles.db')
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS vehicles (
            Number TEXT PRIMARY KEY,
            Img TEXT,
            Status TEXT
        )
    ''')
    conn.commit()
    conn.close()
```

```python
def add_vehicle(number, img, status=None):
    conn = sqlite3.connect('vehicles.db')
    c = conn.cursor()
    c.execute('''
        INSERT OR IGNORE INTO vehicles (Number, Img, Status)
        VALUES (?, ?, ?)
    ''', (number, img, status))
    conn.commit()
    conn.close()
```

```python
import sqlite3
from datetime import datetime, timedelta

# Sample data with DateTime
sample_data = [
    ("ABC123", "images/1.jpg", "Approved", datetime.now() - timedelta(days=1)),
    ("XYZ789", "images/2.jpg", None, datetime.now() - timedelta(days=2)),
    ("DEF456", "images/3.jpg", None, datetime.now() - timedelta(days=3)),
    ("GHI012", "images/4.jpg", None, datetime.now() - timedelta(days=4)),
    ("DEF458", "images/5.jpg", None, datetime.now() - timedelta(days=5))
]

# Adding sample data to the database (if not already added)
def add_sample_data():
    conn = sqlite3.connect('vehicles.db')
    c = conn.cursor()
    
    for number, img, status, dt in sample_data:
        c.execute('''
            INSERT INTO vehicles (Number, Img, Status, DateTime)
            VALUES (?, ?, ?, ?)
        ''', (number, img, status, dt.strftime("%Y-%m-%d %H:%M:%S")))
    
    conn.commit()
    conn.close()

# Run the function to add sample data
add_sample_data()
```

```python
def get_pending_vehicles():
    conn = sqlite3.connect('vehicles.db')
    c = conn.cursor()
    c.execute('SELECT * FROM vehicles WHERE Status IS NULL')
    rows = c.fetchall()
    conn.close()
    return rows
```

```python
import os
import sqlite3
import time
from datetime import datetime
import pandas as pd

# Sample DataFrame with numbers
data = {'Number': ['ABC123', 'XYZ789', 'DEF456']}
df_numbers = pd.DataFrame(data)
print("1")
# Directory to watch
image_directory = cwd
print("2")
# Function to update database with new images
def update_database_with_images():
    conn = sqlite3.connect('/Users/raghavsuri/Desktop/ANPRPM/vehicles.db')
    c = conn.cursor()
    print("3")
    # Get a list of images in the directory
    images = [f for f in os.listdir(image_directory) if os.path.isfile(os.path.join(image_directory, f))]

    for image in images:
        # Extract the number and datetime from the filename
        
        parts = os.path.splitext(image)[0].split('__')
        print(parts)
        if len(parts) != 2:
            continue  
        print("4")
        number, dt_str = parts
        try:
            dt = datetime.strptime(dt_str, "%Y%m%d%H%M%S").strftime("%Y-%m-%d %H:%M:%S")
        except ValueError:
            continue  # Skip if datetime parsing fails

        img_path = os.path.join(image_directory, image)
        print("img_path")
        status = "Approved" if number in df_numbers['Number'].values else None

        # Check if the number and datetime already exist in the database
        c.execute('SELECT COUNT(*) FROM vehicles WHERE Number=? AND DateTime=?', (number, dt))
        if c.fetchone()[0] == 0:
            # Insert into the database
            c.execute('''
                INSERT INTO vehicles (Number, Img, Status, DateTime)
                VALUES (?, ?, ?, ?)
            ''', (number, img_path, status, dt))
            print(f'Inserted {number} with datetime {dt} into database.')

    conn.commit()
    conn.close()

# Run the function every second
while True:
    update_database_with_images()
    time.sleep(1)
```
